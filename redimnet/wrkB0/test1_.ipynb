```python
# plot historgram of audio file 
import matplotlib.pyplot as plt
import numpy as np
import librosa
def plot_histogram(test_audio_file):
    audio, sr = librosa.load(test_audio_file, sr=None)
    plt.figure(figsize=(10, 4))
    plt.hist(audio, bins=50, color='blue', alpha=0.7)
    plt.title('Histogram of Audio Signal')
    plt.xlabel('Amplitude')
    plt.ylabel('Frequency')
    plt.grid()
    plt.show()
# plot histogram of audio file
plot_histogram('./audio/testRob1.wav')
```
